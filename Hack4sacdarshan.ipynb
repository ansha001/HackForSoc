```python
types = ['LIN', 'LM', 'SP', 'BK', 'MU', 'NTER', 'NTRA', 'NAT']
```

```python
no_of_circle = [0]*8
```

```python

for j in range(19):
  for i in range(8):
    no_of_circle[i]=df1.loc[j, types[i]]

  max_value = max(no_of_circle)
  index1 = no_of_circle.index(max_value)
  no_of_circle[index1]=0

  max_value2 = max(no_of_circle)
  index2 = no_of_circle.index(max_value2)
  no_of_circle[index2]=0

  max_value3 = max(no_of_circle)
  index3 = no_of_circle.index(max_value3)

  if max_value>=7 and max_value<=10:
    row = 0
  elif max_value<7 and max_value>=4:
      row = 1
  else:
      row = 2

  if max_value2>=7 and max_value2<=10:
      row2 = 0
  elif max_value2<7 and max_value2>=4:
      row2 = 1
  else:
      row2 = 2

  if max_value3>=7 and max_value3<=10:
      row3 = 0
  elif max_value3<7 and max_value3>=4:
      row3 = 1
  else:
      row3 = 2

  maxtype = types[index1]
  maxtype2 = types[index2]
  maxtype3 = types[index3]

  print("Student ", j+1)
  print(df.loc[row, maxtype])
  print(df.loc[row2, maxtype2])
  print(df.loc[row3, maxtype3])
  print("\n ")



```
